```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("dados_matriculas.xlsx")
```

```python
# grafico de matriculas mensais

plt.figure(figsize=(10, 5))#determino o tamanho 
df['Matrículas'].plot(marker='o', linestyle='-', color='red', mfc='yellow')# determina alguns paramentos, o marker determina como vai ser o marcador, o linestyle determina como vai ser o estilo da linha, e o color deteremina a cor. e o df e depois os [ ] determina de onde eu quero tirar esses dados 
plt.title('Matriculas Mensais')# cria o titulo
plt.xlabel('Mês')#faz o eixo x
plt.ylabel('Número de Matrículas')#faz o eixo y
plt.grid(True)#adiciona linhas
plt.show()
```

```python
# identificar o mes com o maior numero de matricukas

maior_matricula = df['Matrículas'].max()
idice_maior_matricula = df['Matrículas'].idxmax()

print(maior_matricula)
print(idice_maior_matricula)
```

```python
# identificar os mes com o menor numero de matriculas
menor_matricula = df['Matrículas'].min()
idice_menor_matricula = df['Matrículas'].idxmin()

print(menor_matricula)
print(idice_menor_matricula)
```

```python
plt.figure(figsize=(10,5))
df['Matrículas'].plot(marker='s', linestyle='--', color='red', label='Matrículas')

#informacao de destaque
plt.axhline(y=maior_matricula, color='green', linestyle=':', linewidth=2, label=f"Máximo:{maior_matricula}")

plt.axhline(y=menor_matricula, color='purple', linestyle=':', linewidth=2, label=f"Mínimo:{menor_matricula}")

plt.title('Matrículas Mensais')
plt.xlabel('Mês')
plt.ylabel('Número Matriculas')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# calculo medio de matriculas mensais

media_matricula = df['Matrículas'].mean()
print(media_matricula)
```

```python
#plotar o grafico das matriculas mensais
plt.figure(figsize=(10,5))
df['Matrículas'].plot(marker='o', linestyle='-', color='blue', label='matriculas')

#linha de destaque
plt.axhline(y=media_matricula, color='orange', linestyle='--', linewidth=2, label=f"Média:{media_matricula:.2f}")

plt.title("Matrículas mensais")
plt.xlabel("Mês")
plt.ylabel("Número das Matríulas")
plt.grid(True)
plt.legend()
plt.show()

```

```python
#calcular as medias da matricula por trimestre

#passo 1. converter a colua para datetime
df["Mês"] = pd.to_datetime(df["Mês"])

#passo 2. definir a coluna de datas como indice
df.set_index(df["Mês"], inplace=True)

#passo 3. calcular a media do trimesnte
#QE / 4
#QS / semestre
#Q  / quadrimestre
#A / anual
#D / diario
#W / semanal
#H / horario
media_trimestral = df['Matrículas'].resample('QE').mean()
print(media_trimestral)
```

```python
plt.figure(figsize=(10,5))
media_trimestral.plot(kind='bar', color='green')
plt.title("Média de matrículas trimestrais")
plt.xlabel("Trimestre")
plt.ylabel("Média der matriculas")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# calcular variacao percentual mensal
#variacao = df['Matrículas'].pct_change()

#calcular a variacao percentual mensal
variacao = df['Matrículas'].pct_change()*100
```

```python
import seaborn as sn

tabela_variacao = pd.DataFrame({
    'Matricula': df['Matrículas'],
    'Variação Percentual %':variacao
})

def colorir_variacaco(valor):
    if valor > 0:
        color = 'green'
    elif valor < 0:
        color = 'red'
    else:
        color='blue'
    return F'color:{color}'

tabela_variacao.style.map(colorir_variacaco, subset=['Variação Percentual %'])
```
